## 1. How the coefficients get affected?

```python
from sklearn.datasets import load_diabetes

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
data = load_diabetes()
```

```python
df = pd.DataFrame(data.data,columns=data.feature_names)
df['TARGET'] = data.target
```

```python
df.head()
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(data.data,data.target,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import r2_score
```

```python
coefs=[]
r2_scores=[]

#will apply different lambda value to see impact on coefficients
for i in[0,10,100,1000]:
  reg = Ridge(alpha=i)
  reg.fit(X_train,y_train)

  coefs.append(reg.coef_.tolist())
  y_pred=reg.predict(X_test)
  r2_scores.append(r2_score(y_test,y_pred))
```

```python
plt.figure(figsize=(14,9))
plt.subplot(221)
plt.bar(data.feature_names,coefs[0])
plt.title('Alpha= 0, r2_score={}'.format(round(r2_scores[0],2)))


plt.subplot(222)
plt.bar(data.feature_names,coefs[1])
plt.title('Alpha = 10 ,r2_score = {}'.format(round(r2_scores[1],2)))

plt.subplot(223)
plt.bar(data.feature_names,coefs[2])
plt.title('Alpha = 100 ,r2_score = {}'.format(round(r2_scores[2],2)))

plt.subplot(224)
plt.bar(data.feature_names,coefs[3])
plt.title('Alpha = 1000 ,r2_score = {}'.format(round(r2_scores[3],2)))

plt.show()
```

### - In graph can see when "alpha - 0", coefficient range - (-750 to 750)
### - when "alpha - 10" , coefficient range - (-40 to 60)
### - when "alpha - 100" , coefficient range - (-4 to 8)
###-  when "alpha - 1000" , coefficient range - (-0.4 to 0.8)
###- so can see after increasing lambda value to 1000 all coefficient value comes very close to zero.
###- but in ridgre regression coefficient value can never be zero.

3. Impact on Bias and Variance

```python
m = 100
X = 5 * np.random.rand(m, 1) - 2
y = 0.7 * X ** 2 - 2 * X + 3 + np.random.randn(m, 1)

plt.scatter(X, y)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X.reshape(100,1),y.reshape(100),test_size=0.2,random_state=2)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=15)

X_train = poly.fit_transform(X_train)
X_test = poly.transform(X_test)
```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import r2_score

from mlxtend.evaluate import bias_variance_decomp

alpha = np.linspace(0,20,100)

loss = []
bias = []
variance = []

for i in alpha:
    reg = Ridge(alpha=i)
    avg_expected_loss, avg_bias, avg_var = bias_variance_decomp(
        reg, X_train, y_train, X_test, y_test,
        loss='mse',
        random_seed=123)
    loss.append(avg_expected_loss)
    bias.append(avg_bias)
    variance.append(avg_var)
```

```python
plt.plot(alpha,loss,label='loss')
plt.plot(alpha,bias,label='Bias')
plt.plot(alpha,variance,label='Variance')
plt.ylim(0,5)
plt.xlabel('Alpha')
plt.legend()
```

```python
from sklearn.datasets import make_regression

X,y = make_regression(n_samples=100, n_features=1, n_informative=1, n_targets=1,noise=20,random_state=13)

plt.scatter(X,y)

from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(X,y)
print(reg.coef_)
print(reg.intercept_)
```

```python
def cal_loss(m,alpha):
  return np.sum((y - m*X.ravel() + 2.29)**2)+ alpha*m*m

```
